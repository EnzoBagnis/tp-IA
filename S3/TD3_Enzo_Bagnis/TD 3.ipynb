# Exercice 0

## Q1

```python
import numpy as np
import matplotlib.pyplot as plt
```

## Q2

```python
from sklearn import datasets
```

## Q3

```python
iris = datasets.load_iris()
```

## Q4

```python
iris
```

# Exercice 1

# Q1

```python
x = iris.data[:, :2]
```

# Q2

```python
y = iris.target.copy()
y[y != 0] = 1

```

# Exercice 2

```python
plt.figure()
plt.scatter(x[y == 0][:, 0], x[y == 0][:, 1], color='green', label='Class 0')
plt.scatter(x[y == 1][:, 0], x[y == 1][:, 1], color='yellow', label='Class 1')
plt.xlabel('Sepal length')
plt.ylabel('Sepal width')
plt.legend()
plt.show()
```

# Exercice 3

# Q1

```python
from sklearn.linear_model import LogisticRegression
```

# Q2

```python
model = LogisticRegression(C=1e20)
```

## Q3

```python
model.fit(x, y)
```

# Exercice 4

# Q1

```python
Iries_to_Predict = [[5.5, 2.5], [7, 3], [3, 2], [5, 3]]
```

# Q2

```python
Iries_np = np.array(Iries_to_Predict)
plt.figure()
plt.scatter(x[y == 0][:, 0], x[y == 0][:, 1], color='green', label='Class 0')
plt.scatter(x[y == 1][:, 0], x[y == 1][:, 1], color='yellow', label='Class 1')
plt.scatter(Iries_np[:, 0], Iries_np[:, 1], color='red', label='To Predict', s=100, marker='x')
plt.xlabel('Sepal length')
plt.ylabel('Sepal width')
plt.legend()
plt.show()
```

# Q3

```python
predictions = model.predict(Iries_to_Predict)
print(predictions)
```

# Q4

```python
for point, pred in zip(Iries_to_Predict, predictions):
    print(f"Point {point} -> Class {pred}")
```

# Exercice 5

# Q1

```python
y_q1 = iris.target.copy()
y_q1[y_q1 == 2] = 0

model_q1 = LogisticRegression(C=1e20)
model_q1.fit(x, y_q1)

plt.figure()
plt.scatter(x[y_q1 == 0][:, 0], x[y_q1 == 0][:, 1], color='green', label='Class 0 (Rest)')
plt.scatter(x[y_q1 == 1][:, 0], x[y_q1 == 1][:, 1], color='yellow', label='Class 1 (Versicolor)')
plt.scatter(Iries_np[:, 0], Iries_np[:, 1], color='red', label='To Predict', s=100, marker='x')
plt.xlabel('Sepal length')
plt.ylabel('Sepal width')
plt.legend()
plt.title('Versicolor vs Rest')
plt.show()

predictions_q1 = model_q1.predict(Iries_to_Predict)
for point, pred in zip(Iries_to_Predict, predictions_q1):
    print(f"Point {point} -> Class {pred}")
```

# Q2

```python
y_q2 = iris.target.copy()
y_q2[y_q2 == 0] = 1
y_q2[y_q2 == 2] = 0

model_q2 = LogisticRegression(C=1e20)
model_q2.fit(x, y_q2)

plt.figure()
plt.scatter(x[y_q2 == 0][:, 0], x[y_q2 == 0][:, 1], color='green', label='Class 0 (Virginica)')
plt.scatter(x[y_q2 == 1][:, 0], x[y_q2 == 1][:, 1], color='yellow', label='Class 1 (Rest)')
plt.scatter(Iries_np[:, 0], Iries_np[:, 1], color='red', label='To Predict', s=100, marker='x')
plt.xlabel('Sepal length')
plt.ylabel('Sepal width')
plt.legend()
plt.title('Virginica vs Rest')
plt.show()

predictions_q2 = model_q2.predict(Iries_to_Predict)
for point, pred in zip(Iries_to_Predict, predictions_q2):
    print(f"Point {point} -> Class {pred}")
```
