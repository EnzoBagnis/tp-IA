## Exercice 0

```python
import pandas as pd

# 2. Lecture du fichier avec Pandas
# sep='\s+' permet de gérer un ou plusieurs espaces comme séparateur
df = pd.read_csv('donnees.dat', sep=r'\s+', header=None)

print("--- Contenu du DataFrame ---")
print(df)


x = df[0].tolist()
y = df[1].tolist()

print("\n--- Listes générées ---")
print("Liste x :", x)
print("Liste y :", y)
```

# Exercice 1

```python
x_somme = sum(x)
y_somme = sum(y)
x_carré_somme = sum([i**2 for i in x])
xy_somme = sum([x[i]*y[i] for i in range(len(x))])

# Calcul de N (nombre de points)
N = len(x)

# Calcul du dénominateur commun
denominateur = N * x_carré_somme - x_somme**2

# Calcul de a et b
a = (N * xy_somme - x_somme * y_somme) / denominateur
b = (x_carré_somme * y_somme - x_somme * xy_somme) / denominateur

print(f"N = {N}")
print(f"a = {a}")
print(f"b = {b}")
```

# Exercice 2

```python
def regression_lineaire(x, y):

    N = len(x)
    x_somme = sum(x)
    y_somme = sum(y)
    x_carre_somme = sum([i**2 for i in x])
    xy_somme = sum([x[i]*y[i] for i in range(N)])

    denominateur = N * x_carre_somme - x_somme**2

    # Calcul de a et b
    a = (N * xy_somme - x_somme * y_somme) / denominateur
    b = (x_carre_somme * y_somme - x_somme * xy_somme) / denominateur

    return a, b

# Appel de la fonction avec les listes x et y précédemment créées
a_res, b_res = regression_lineaire(x, y)

# Affichage avec 3 décimales
print(f"a = {a_res:.3f}")
print(f"b = {b_res:.3f}")
```

# Exercice 3

```python
import numpy as np

# Conversion des listes en tableaux Numpy
x_np = np.array(x)
y_np = np.array(y)

def regLin_np(x, y):
    """
    Calcule les paramètres a et b de la régression linéaire avec Numpy.

    Args:
        x (numpy.ndarray): Tableau des valeurs de x.
        y (numpy.ndarray): Tableau des valeurs de y.

    Returns:
        tuple: (a, b)
    """
    N = len(x)
    # Utilisation de la méthode sum() de Numpy et des opérations vectorielles
    x_somme = np.sum(x)
    y_somme = np.sum(y)
    # Calcul de la somme des carrés et des produits directement avec les opérations vectorielles
    x_carre_somme = np.sum(x**2)
    xy_somme = np.sum(x * y)

    denominateur = N * x_carre_somme - x_somme**2

    a = (N * xy_somme - x_somme * y_somme) / denominateur
    b = (x_carre_somme * y_somme - x_somme * xy_somme) / denominateur

    return a, b

# Appel de la fonction Numpy
a_np, b_np = regLin_np(x_np, y_np)

print(f"Avec Numpy : a = {a_np:.3f}, b = {b_np:.3f}")

# Vérification que regLin_np et la fonction précédente (regression_lineaire) renvoient les mêmes résultats
print(f"Différence sur a : {abs(a_res - a_np)}")
print(f"Différence sur b : {abs(b_res - b_np)}")

if np.isclose(a_res, a_np) and np.isclose(b_res, b_np):
    print("Les résultats sont identiques.")
else:
    print("Les résultats sont différents.")
```
